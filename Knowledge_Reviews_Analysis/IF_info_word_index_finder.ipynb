```python
import pandas as pd
import re
import nltk
from nltk import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
df = pd.read_excel('/Users/angieryu2202/Desktop/IF_App_Study/sents_df.xlsx')
```

```python
df.info()
```

```python
words = []
positions = []
find_words = ['article', 'tip', 'information', 'guidance', 'insight', 'learn', 'knowledge', 'faq', 'question', 'post', 'reading', 'pointer']
for sent in df['sents']:
    sent = re.sub(r'[^\w\s]','',sent).lower()
    print(sent)
    lemmer = WordNetLemmatizer()
    lemmatized = [lemmer.lemmatize(w, 'n') for w in word_tokenize(sent)]
    for word in find_words:
        if word in lemmatized:
            words.append(word)
            print(word)
            #result = [i+1 for i, w in enumerate(sent.split('')) if w.lower == word]
            #print(result)  
            #start_index = sent.find(word)
            #end_index = start_index + len(word) # if the start_index is not -1
            #position = end_index/len(sent)
            position = (lemmatized.index(word)+1)/len(lemmatized) # 1   index 0  
            print(lemmatized.index(word)+1)
            print(len(lemmatized))
            positions.append(position)
            print(position) #     
```

```python
#!pip3 install scipy
from scipy.stats import skew
import matplotlib.pyplot as plt
import seaborn as sns
index_avgs = []
skews = []
stan_devs = []
def index_avg_calculator(keyword):
    globals()[keyword+'_df'] = index_df[index_df['words'].str.contains(keyword)]
    print(globals()[keyword+'_df'].describe())
    index_avg = globals()[keyword+'_df']['positions'].mean()
    stan_dev = globals()[keyword+'_df']['positions'].std()
    #print(keyword, " : ", str(index_avg)+"("+str(stan_dev)+")")
    skewness =  skew(globals()[keyword+'_df']['positions'])
    if skewness == 0:
        skews.append('normal')
    elif skewness > 0:
        skews.append('positive') # long right tail (    - tail     )
    elif skewness < 0:
        skews.append('negative') # long left tail (   )
    index_avgs.append(index_avg)
    stan_devs.append(stan_dev)
    #skews.append(skewness)
```
